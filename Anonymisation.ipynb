## Générer des données

Dans ce qui suit nous allons essayer de générer un jeu de données que nous allons anonymiser ensuite

```python
import csv
import random
import uuid
import codecs
import io
from collections import OrderedDict
import requests

def generateUUID() :
    uuidOne = uuid.uuid1()
    return uuidOne
def generateRandomLatLngFile(minLat, maxLat, minLng, maxLng, numberOfRandomLatLng):
    with open('randomLatLng.csv', 'w', newline='') as file:
        writer = csv.writer(file, delimiter=";") 
        writer.writerow(["lat", "lng"])  
        for i in range(numberOfRandomLatLng):
            randomLat = random.uniform(minLat, maxLat)
            randomLng = random.uniform(minLng, maxLng)
            writer.writerow([randomLat, randomLng]) 
def convertLatLngToRealAddress(fileName):
    requests_session = requests.Session()
    kwargs = {
        'data': OrderedDict([
            ('data', (fileName,
             io.BytesIO(
                 open(fileName, 'rb').read()
             )))
        ]),
        'method': 'post',
        'files': OrderedDict([
            ('data', (fileName,
             io.BytesIO(
                 open(fileName, 'rb').read()
             )))
        ]),
        'stream': True,
        'url': 'https://api-adresse.data.gouv.fr/reverse/csv/'
    }
    
    response = requests_session.request(**kwargs)
    
    with codecs.open('reversed-'+fileName, 'wb', 'utf-8') as f:
        f.write(response.text)
    


def generateRandomAdultAge():
    age = random.randint(18,100)
    return age

def generateRandomFirstAndLastName():
    first_names=("James", "Robert", "John", "Michael", "David", "William", "Richard", 
             "Joseph", "Thomas", "Charles", "Christopher", "Daniel", "Matthew",
             "Anthony", "Mark", "Donald", "Steven", "Paul", "Andrew", "Joshua",
             "Kenneth", "Kevin", "Brian", "George", "Timothy", "Ronald", "Edward",
             "Jason", "Jeffrey", "Ryan", "Jacob", "Gary", "Nicholas", "Eric",
             "Jonathan", "Stephen", "Larry", "Justin", "Scott", "Brandon", "Benjamin",
             "Samuel", "Gregory", "Alexander", "Frank", "Patrick", "Raymond", "Jack",
             "Dennis", "Jerry", "Tyler", "Aaron", "Jose", "Adam", "Nathan", "Henry",
             "Douglas", "Zachary", "Peter", "Kyle", "Ethan", "Walter", "Noah",
             "Jeremy", "Christian", "Keith", "Roger", "Terry", "Gerald", "Harold",
             "Sean", "Austin", "Carl", "Arthur", "Lawrence", "Dylan", "Jesse",
             "Jordan", "Bryan", "Billy", "Joe", "Bruce", "Gabriel", "Logan", "Albert",
             "Willie", "Alan", "Juan", "Wayne", "Elijah", "Randy", "Roy", "Vincent",
             "Ralph", "Eugene", "Russell", "Bobby", "Mason", "Philip", "Louis")
    
    last_names=("SMITH", "JOHNSON", "WILLIAMS", "BROWN", "JONES", "GARCIA", "MILLER",
            "DAVIS", "RODRIGUEZ", "MARTINEZ", "HERNANDEZ", "LOPEZ", "GONZALEZ",
            "WILSON", "ANDERSON", "THOMAS", "TAYLOR", "MOORE", "JACKSON", "MARTIN",
            "LEE", "PEREZ", "THOMPSON", "WHITE", "HARRIS", "SANCHEZ", "CLARK",
            "RAMIREZ", "LEWIS", "ROBINSON", "WALKER", "YOUNG", "ALLEN", "KING",
            "WRIGHT", "SCOTT", "TORRES", "NGUYEN", "HILL", "FLORES", "GREEN", "ADAMS",
            "NELSON", "BAKER", "HALL", "RIVERA", "CAMPBELL", "MITCHELL", "CARTER",
            "ROBERTS", "GOMEZ", "PHILLIPS", "EVANS", "TURNER", "DIAZ", "PARKER",
            "CRUZ", "EDWARDS", "COLLINS", "REYES", "STEWART", "MORRIS", "MORALES",
            "MURPHY", "COOK", "ROGERS", "GUTIERREZ", "ORTIZ", "MORGAN", "COOPER",
            "PETERSON", "BAILEY", "REED", "KELLY", "HOWARD", "RAMOS", "KIM", "COX",
            "WARD", "RICHARDSON", "WATSON", "BROOKS", "CHAVEZ", "WOOD", "JAMES",
            "BENNETT", "GRAY", "MENDOZA", "RUIZ", "HUGHES", "PRICE", "ALVAREZ",
            "CASTILLO", "SANDERS", "PATEL", "MYERS", "LONG", "ROSS", "FOSTER", "JIMENEZ")
    firstAndLastName = random.choice(first_names)+" "+random.choice(last_names)
    return firstAndLastName

def generateRandomCreditCardNumber():
    creditCardNumber = random.randint(1000000000000000,9999999999999999)
    return creditCardNumber

def generateRandomMobileNumber() :
    mobileNumber = random.randint(33600000000,33699999999)
    return mobileNumber

def generateRandomNumberOfOrders():
    numberOfOrders = random.randint(0,40)
    return numberOfOrders

def generateRandomList(numberOfLines):
    listOfAddresses = None
    with open("reversed-randomLatLng.csv", 'r', encoding='utf-8') as file:
        csvreader = csv.reader(file, delimiter = ";")
        next(csvreader)
        listOfAddresses = list(csvreader)
    
    lines = [["id", "name", "street", "postalCode", "city", "age", "creditCardNumber", "mobileNumber", "numberOfOrders"]]
    for i in range(numberOfLines) :
        # do something here
        uuid = generateUUID()
        name = generateRandomFirstAndLastName()
        street = listOfAddresses[i][9]
        postalCode = listOfAddresses[i][11]
        city = listOfAddresses[i][12]
        age = generateRandomAdultAge()
        creditCardNumber = generateRandomCreditCardNumber()
        mobileNumber = generateRandomMobileNumber()
        numberOfOrders = generateRandomNumberOfOrders()
        lines.append([uuid, name, street, postalCode, city, age, creditCardNumber, mobileNumber, numberOfOrders])
    return lines



numberOfLines = 1000
generateRandomLatLngFile(48.811649, 48.903132, 2.252369, 2.424717, numberOfLines)
convertLatLngToRealAddress("randomLatLng.csv")

randomList = generateRandomList(numberOfLines)
with open('customers.csv', 'w', newline='') as file:
    writer = csv.writer(file, delimiter=";") 
    for thisLine in randomList :
        writer.writerow(thisLine)
```
